```python
import numpy as np
import matplotlib.pyplot as plt

# x-axis
x = np.linspace(0, 12, 2000)

# transparent figure
fig, ax = plt.subplots(figsize=(14, 2.5))
fig.patch.set_alpha(0.0)
ax.set_facecolor("none")

# draw many similar waves
n_lines = 18
for i in range(n_lines):
    shift = i * 0.18
    amp   = 1.0 - i * 0.02

    y = (
        0.9 * np.sin(1.6 * x + shift) +
        0.5 * np.cos(2.7 * x - 0.7 * shift) +
        0.25 * np.sin(0.7 * x + 1.2 * shift)
    ) * amp

    ax.plot(x, y, color="white",  lw=1.2)

# remove all axes
ax.axis("off")
plt.margins(0)
plt.tight_layout(pad=0)

# save with transparent background
plt.savefig("transparent_waves.png", dpi=300, transparent=True, bbox_inches="tight", pad_inches=0)
plt.show()
```
